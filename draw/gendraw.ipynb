```python
# this code can be only used for plotting
import pandas as pd
import matplotlib.pyplot as plt
#read in the data we use, and get the number of lines
N=5
data=[0 for i in range(N)]
data[0] = pd.read_excel('F-PWA93.xlsx', engine='openpyxl')
data[1]=  pd.read_excel('F-lo(LS).xlsx', engine='openpyxl')
data[2]=  pd.read_excel('F-lo500(LS).xlsx', engine='openpyxl')
data[3]=  pd.read_excel('F-lo(pt).xlsx', engine='openpyxl')
data[4]=  pd.read_excel('F-lo500(pt).xlsx', engine='openpyxl')
# 获取 x 值
x_values = data[0]['T']

# 获取 y 值
y=[[0 for col in range(4)] for row in range(N)]
for i in range(N):
    y[i][0] = data[i]['1F3']
    y[i][1] = data[i]['3F2']
    y[i][2] = data[i]['3F3']
    y[i][3] = data[i]['3F4']
fig=plt.figure()
ax=[0 for i in range(4)]
for i in range(4):
    ax[i] = fig.add_subplot(2, 2,i+1)
    ax[i].plot(x_values,list(map(float, y[0][i])),'ro',zorder=1)
    ax[i].plot(x_values,list(map(float, y[1][i])),color='b',linestyle='--',zorder=0,label='LS')
    ax[i].plot(x_values,list(map(float, y[2][i])),color='g',linestyle='--',zorder=0,label='LS500')
    ax[i].plot(x_values,list(map(float, y[3][i])),color='y',linestyle=':',zorder=0,label='pt')
    ax[i].plot(x_values,list(map(float, y[4][i])),color='c',linestyle=':',zorder=0,label='pt500')
font1={'family':'Times New Roman',
'weight':'normal',
'size':20,
}
ax[0].text(200,-1,'$\mathregular{^1F_3}$',font1)
ax[1].text(100,2,'$\mathregular{^3F_2}$',font1)
ax[2].text(200,-1,'$\mathregular{^3F_3}$',font1)
ax[3].text(100,2,'$\mathregular{^3F_4}$',font1)
plt.legend()
plt.show()
```
